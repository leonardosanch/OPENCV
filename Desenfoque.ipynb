```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread('ladrillos.jpg').astype(np.float32)/255
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gamma = 2
gamma_img = np.power(img, gamma)
plt.imshow(gamma_img)
```

```python

fuente = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text='Hola', org=(10,300),fontFace=fuente, fontScale = 7, color=(255,0,0), thickness=3)
plt.imshow(img)
#plt.show()
```

```python
kernel = np.ones(shape=(5,5), dtype=np.float32)/25
```

```python
kernel
```

```python
img_k=cv2.filter2D(img,-1,kernel)
```

```python
plt.imshow(img_k)
```

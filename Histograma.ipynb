```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
perro = cv2.imread('perro.jpg')
mostrar_perro = cv2.cvtColor(perro, cv2.COLOR_BGR2RGB)

ladrillos = cv2.imread('ladrillos.jpg')
mostrar_ladrillos = cv2.cvtColor(ladrillos, cv2.COLOR_BGR2RGB)

crucigrama = cv2.imread('crucigrama.jpg')
mostrar_crucigrama = cv2.cvtColor(crucigrama, cv2.COLOR_BGR2RGB)

```

```python
plt.imshow(mostrar_perro)
```

```python
#OPENCV BGR
hist = cv2.calcHist([perro], channels=[2], mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist)
```

```python
#OPENCV BGR
hist = cv2.calcHist([ladrillos], channels=[2], mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist)
```

```python
plt.imshow(mostrar_ladrillos)
```

```python
img = perro
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('perro.jpg')
plt.show()
```
